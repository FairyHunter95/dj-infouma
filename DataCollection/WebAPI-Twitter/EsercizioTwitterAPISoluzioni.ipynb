# Domanda 2
Per ogni follower salvare l'elenco dei tweets nella lista tweets. Per ogni tweet salvare l'autore, il testo e la data di creazione. Gestire le eccezioni utilizzando il blocco try - except.

```python
from twitter import TwitterError
tweets = []
# visto che la lista è troppo lunga, considero solo i primi 5 followers
for follower in followers[:5]:
    try:
        screen_name = follower.screen_name
        timeline = api.GetUserTimeline(screen_name=screen_name)
        for tweet in timeline:
            line = {'author' : screen_name, 'text' : tweet.text, 'date' : tweet.created_at}
            tweets.append(line)
    except TwitterError:
        continue
print(tweets)
```

# Domanda 3
Salvare il contenuto della lista tweets su un CSV. Verificare la correttezza del file salvato. (Si trova nella cartella in cui è aperto il notebook).

```python
import csv

keys = tweets[0].keys()
with open('tweets.csv', 'w') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(tweets)
```
